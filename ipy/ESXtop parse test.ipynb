```python
import csv
import json
import os
import pandas as pd
```

```python
#Function - REad the csv File
def readcsv(filename, delim=','):
	reader = csv.reader(open(filename,'rU'), delimiter=delim)
	csvdata=[]
	for row in reader:
		csvdata.append(row)
	return csvdata
```

```python
# Function - returns a header row for the table
def get_header_row(csvdata, index=0):
	headers={'names':[], 'cols':{}, 'count':0}
	poss_headers=csvdata[index]
	#col_count=0
	#subbing in enumberate
	for col_count,data in list(enumerate(poss_headers)):								#For each number row and value
		if (len(data)>5) and (type(data)==type('hello') and ' ' in data):
			headers['names'].append(data)
			headers['cols'][col_count] = {'name': data, 'Values':[]}
	headers.update({'count':col_count})
	return headers
```

```python
# function - parse headers for content
def parse_header_for_content(headers, parse_on=[]):
	temp_dict={'phrases':{}}
	for x in parse_on:					#Initialize the phrase dict
		temp_dict['phrases'][x]=[]
	#temp_dict['phrases'][x]=[] for x in parse_on			#tried one liner	
	for i in range(len(headers['names'])):
		for j in parse_on:
			if j.lower() in headers['names'][i].lower():
				temp_dict['phrases'][j].append(i)
	return temp_dict	
```
